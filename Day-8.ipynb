```python
def sum_of_elements(lst):
    total=0
    for i in range(len(lst)):
        total+=lst[i]
    return total
lst=[1,2,3,4,5,6]
print(sum_of_elements(lst))
```

```python
def calculator(x, y, op):
    if op == "add":
        return x + y
    elif op == "sub":
        return x - y
    elif op == "mul":
        return x * y
    elif op == "mod":
        return x % y
    else:
        return "Invalid operation"

n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))

print("Addition:", calculator(n1, n2, "add"))
print("Subtraction:", calculator(n1, n2, "sub"))
print("Multiplication:", calculator(n1, n2, "mul"))
print("Modulus:", calculator(n1, n2, "mod"))
```
