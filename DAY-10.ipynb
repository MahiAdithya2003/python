```python
l=[1,2,3,4,5]
l.append(10)
print(l)
l.extend([10,20,30])
print(l)
l.insert(3,100)
print(l)
l.insert(100,1000)
print(l)
l.insert(-1,200)
print(l)
```

```python
#removing elements from list 
#remove(val):it removes the first occurence value from list ,otherwise value error 
#pop(pos):It removes and return end value from list,if list is empty or raise index error
pop(position):it removes and return specificed position element from list
#clear():it removes all elements fro list.
```

```python
l=[1,2,3,4,5]
l.remove(3)
print(l)
val=l.pop()
print(val)
print(l)
print(l.pop(2),1)
l.clear()
print(l)
```

```python
present = 0
absent = 0
i = 1

while i <= 10:
    x = input("Student " + str(i) + " (present/absent): ")

    if x == 'p':
        present += 1
    else:
        absent += 1

    i += 1

print("Present:", present)
print("Absent:", absent)
```
